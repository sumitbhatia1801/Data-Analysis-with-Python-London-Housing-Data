## Convert the datatype of "date" column to date-time_format.

```python
data.head()
```

#### Answer - Job done

## Add a column year in the dataframe, which contains years only

```python
data.head()
```

```python
data["year"] = data["date"].dt.year
```

```python
data.head()
```

#### Answer - Job done

## Add a new column "month" as 2nd column in the data frame, which contains month only.

```python
data.head()
```

```python
data.insert(loc = 1 , column = "month" , value = data["date"].dt.month)
```

```python
data.head()
```

#### Answer - Job Done

## Remove the columns "year" and "month" from the dataframe.

```python
data.drop(columns = ["year" , "month"] , axis = 1 , inplace = True)
```

```python
data.head()
```

#### Answer - Job done
